```python
from unsloth import FastModel
import torch

model_name = "unsloth/gemma-3n-E4B-it"
model, tokenizer = FastModel.from_pretrained(
    model_name = model_name,
    dtype = torch.bfloat16, #it tieu hao bo nho
    max_seq_length = 1024,
    load_in_4bit = True
)
```

```python
prompt = [
    {
        "role":"system",
        "content":[{
            "type":"text",
            "text":"Bạn là 1 trợ lí AI hữu ích, hãy trả lời câu hỏi 1 cách chính xác"
        }]
    },
    {
        "role":"user",
        "content":[{
            "type":"text",
            "text":"Viết hàm tính tổng 2 số bằng python."
          }
        ]

    },
]
```

```python
from transformers import TextStreamer

def inference(messages, max_new_tokens = 128):
  return model.generate(
      **tokenizer.apply_chat_template(
          messages,
          add_generation_prompt = True,
          tokenize = True,
          return_dict = True,
         return_tensors = "pt"
      ).to("cuda"),
      max_new_tokens = max_new_tokens,
      temperature = 1.0,
      streamer = TextStreamer(tokenizer,skip_prompt = True)
  )
```

```python
result = inference(prompt)
```

```python
from unsloth.chat_templates import get_chat_template

tokenizer = get_chat_template(
    tokenizer,
    chat_template = "gemma-3"
)
```

```python
from datasets import load_dataset

dataset = load_dataset("Aixr/Math-Question-Answer")
```

```python
from unsloth.chat_templates import standardize_data_formats

dataset_2 = standardize_data_formats(dataset['train'])
```

```python
from unsloth.chat_templates import standardize_data_formats

def format_prompt(question,answer):
  prompt = [
      {
          "role":"system",
          "content":[
              {
                  "type":"text",
                  "text":"You are AI assistant. Please answer user's question exactly"
              }
          ]
      },
      {
        "role":"user",
          "content":[{
               "type":"text",
               "text":question
           }
        ]
      },
      {
          "role":"assistant",
          "content":[
              {
                  "type":"text",
                  "text":answer
              }
          ]

      }

  ]
  return prompt


def formatting_prompts_func(examples):
  n = len(examples['question'])
  texts_2 = []

  # print(problem)
  for i in range(n):

    texts = tokenizer.apply_chat_template(format_prompt(examples['question'][i],examples['answer'][i]),add_generation_prompt=False).removeprefix("<bos>")
    texts_2.append(texts)

  return {"texts": texts_2}

dataset_2 = dataset_2.map(formatting_prompts_func,batched = True)
```

```python
dataset_2['texts'][0]
```

```python
from unsloth import FastModel

model = FastModel.get_peft_model(
    model,
    r = 16,
    lora_alpha = 16,
    lora_dropout = 0,
    bias = "none",

)
```

```python
from trl import SFTTrainer, SFTConfig

trainer = SFTTrainer(
    model = model,
    tokenizer = tokenizer,
    train_dataset = dataset_2,
    eval_dataset = None,
    args = SFTConfig(
        dataset_text_field = "texts",
        per_device_train_batch_size = 1,
        gradient_accumulation_steps = 4,
        warmup_steps = 5,
        max_steps = 60,
        learning_rate = 2e-4,
        logging_steps = 1,
        optim = "adamw_8bit",
        weight_decay = 0.01,
        lr_scheduler_type = "linear",
        seed = 3407,
        report_to = "none"
    ),
    formatting_func = formatting_prompts_func,
)
```

```python
from unsloth.chat_templates import train_on_responses_only
trainer = train_on_responses_only(
    trainer,
    instruction_part = "<start_of_turn>user\n",
    response_part = "<start_of_turn>model\n",
)
```

```python
import torch
torch._dynamo.config.cache_size_limit = 512
history = trainer.train()
```

```python
question = "Find the derivative of the function f(x) = 8x^2 + 2.e^x"
prompt = [
    {
        "role": "system",
        "content": [
            {
                "type": "text",
                "text": "You are AI assistant. Please answer user's question exactly"
            }
        ]
    },
    {
        "role": "user",
        "content": [
            {
                "type": "text",
                "text": question
            }
        ]
    }
]
```

```python
result = inference(prompt)
```
